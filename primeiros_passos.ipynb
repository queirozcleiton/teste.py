```
vendas = [1500, 3200, 800, 4100, 2750]

total = 0
for venda in vendas:
    total = total + venda

print("Total de vendas:", total)
print("Quantidade de meses:", len(vendas))
print("Média mensal:", total / len(vendas))
```

# Funções

```
def calcular_media(valores):
  total = 0
  for valor in valores:
    total = total + valor
  return total / len(valores)

vendas = [1500, 3200, 800, 4100, 2750]
despesas = [900, 1800,600,2200,1100]

print("Média de vendas:", calcular_media(vendas))
print("Média de despesas:", calcular_media(despesas))
```

# Automação de Arquivos

## Carregar Biblioteca e crias Planilha

```
import openpyxl

#Criar uma planilha nova
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Vendas"

#Escrever em células
ws["A1"] = "Mês"
ws["B1"] = "Valor"
ws["A2"] = "Janeiro"
ws["B2"] = 1500

#Salvar
wb.save("minhas_vendas.xlsx")
print("Arquivo criado!")
```

## Funções

```
def calcular_media(valores):
  total = 0
  for valor in valores:
    total = total + valor
  return total / len(valores)

vendas = [1500, 3200, 800, 4100, 2750]
despesas = [900, 1800,600,2200,1100]

print("Média de vendas:", calcular_media(vendas))
print("Média de despesas:", calcular_media(despesas))
```

## Carregar Biblioteca e criar Planilha

```
import openpyxl

#Criar uma planilha nova
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Vendas"

#Escrever em células
ws["A1"] = "Mês"
ws["B1"] = "Valor"
ws["A2"] = "Janeiro"
ws["B2"] = 1500

#Salvar
wb.save("minhas_vendas.xlsx")
print("Arquivo criado!")
```

```
import openpyxl

#Criar uma planilha nova
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Vendas"

#Cabeçalho
ws["A1"] = "Mês"
ws["B1"] = "Valor"

# Dados
meses = [
    ["Janeiro", 1500],
    ["Fevereiro", 3200],
    ["Março", 800],
    ["Abril", 4100],
    ["Maio", 2750]
]

for i, mes in enumerate(meses):
  ws[f"A{i+2}"] = mes[0]
  ws[f"B{i+2}"] = mes[1]

wb.save("vendas_completo.xlsx")
print("Arquivo criado!")
```

## Carregar planilha existente

```
import openpyxl

wb = openpyxl.load_workbook("vendas_completo.xlsx")
ws = wb.active

for linha in ws.iter_rows(min_row=2, values_only=True):
  mes = linha[0]
  valor = linha[1]
  print(f"{mes}: R$ {valor}")
```

## Ler calcular e escrever

```
import openpyxl

wb = openpyxl.load_workbook("vendas_completo.xlsx")
ws = wb.active

#Ler e somar
total = 0
for linha in ws.iter_rows(min_row=2, values_only=True):
  total = total + linha[1]

#Escrever o total de volta na planilha
ultima_linha = ws.max_row + 2
ws[f"A{ultima_linha}"] = "Total"
ws[f"B{ultima_linha}"] = total

#Salvar
wb.save("vendas_completo.xlsx")
print(f"Total R$ {total} inserido na planilha!")
```

```
import openpyxl

#Criar uma planilha nova
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "Resultado"

#Cabeçalho
ws["A1"] = "Mês"
ws["B1"] = "Vendas"
ws["C1"] = "Despesas"

# Dados
meses = [
    ["Janeiro", 1500, 900],
    ["Fevereiro", 3200, 1800],
    ["Março", 800, 600],
    ["Abril", 4100, 2200],
    ["Maio", 2750, 1100]
]

for i, mes in enumerate(meses):
  ws[f"A{i+2}"] = mes[0]
  ws[f"B{i+2}"] = mes[1]
  ws[f"C{i+2}"] = mes[2]

wb.save("relatorio_resultado.xlsx")
print("Arquivo criado!")
```

```
import openpyxl

wb = openpyxl.load_workbook("relatorio_resultado.xlsx")
ws = wb.active

# criar a coluna Lucro
ws["D1"] = "Lucro"

# Calcular e escrever na coluna Lucro
for linha in ws.iter_rows(min_row=2, values_only=True):
  mes = linha[0]
  vendas = linha[1]
  despesas = linha[2]
  lucro = vendas - despesas
  ws[f"D{linha+1}"] = lucro

#Ler e somar
vendas = 0
despesas = 0
lucro = 0
for linha in ws.iter_rows(min_row=2, values_only=True):
  vendas = vendas + linha[1]
  despesas = despesas + linha[2]
  lucro = lucro + linha[3]

#Escrever o total de volta na planilha
ultima_linha = ws.max_row + 2
ws[f"A{ultima_linha}"] = "TOTAL"
ws[f"B{ultima_linha}"] = vendas
ws[f"C{ultima_linha}"] = despesas
ws[f"D{ultima_linha}"] = lucro

#Salvar
wb.save("vendas_completo.xlsx")
print(f"Lucro total R$ {lucro} inserido na planilha!")
```
